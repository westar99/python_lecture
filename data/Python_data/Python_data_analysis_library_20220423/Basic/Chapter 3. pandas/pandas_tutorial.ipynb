# 새로운 데이터 추가

```python
juice['Sold'] = 0
print(juice.tail(3))
```

```python
juice['Sold'] = juice['Lemon'] + juice['Orange']
print(juice.tail(3))
```

```python
juice['Revenue'] = juice['Price'] * juice['Sold']
print(juice.head(3))
```

- drop() 함수 : 특정 행과 열을 제거
- column 제거

```python
juice_col_drop = juice.drop('Sold', axis = 1)
print(juice_col_drop.head(3))
```

```python
juice_col_drop = juice.drop('새로운 컬럼명', axis = 1)
print(juice_col_drop.head(3))
```

```python
juice_row_drop = juice_col_drop.drop(0, axis = 0)
juice_row_drop.head()
```

```python
juice_row_drop.reset_index(drop=True)
```

## loc, iloc를 활용한 추출
- iloc : 인덱싱 기반
- loc : 라벨 기반 ("이름")

```python
juice
```

```python
%%time

juice.iloc[0:3, 0:2]
```

```python
%%time

juice.loc[0:2, ['Date', 'Location']]
```

```python
juice.loc[juice['Revenue'] > 100, ['Date', 'Revenue']]
```

```python
juice.iloc[juice['Revenue'] > 100, 0:2]
```

## Group by
- 전처리의 마지막 단계
- 집계함수 --> 피벗테이블

- Location, 매출액의 최대, 최소, 총합계, 평균 금액을
---> Beach
---> Park

```python
import numpy as np
juice.groupby('Location')[['Revenue', 'Sold', 'Temperature']].agg([sum, np.mean])
```
