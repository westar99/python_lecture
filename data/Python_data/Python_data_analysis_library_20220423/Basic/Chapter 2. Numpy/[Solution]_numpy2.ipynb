# 배열 연산
- 간단한 사칙 연산을 구현한다. 

```python
import numpy as np
array_01 = np.array([1, 2, 3])
array_02 = np.array([10, 20, 30])
```

```python
# 덧셈
newArr = np.add(array_01, array_02)
print("덧셈: ", newArr)
```

```python
# 뺄셈
newArr = np.subtract(array_01, array_02)
print("뺄셈: ", newArr)
```

```python
# 곱셈
newArr = np.multiply(array_01, array_02)
print("곱셈: ", newArr)
```

```python
# 나눗셈
newArr = np.divide(array_01, array_02)
print("나눗셈: ", newArr)
```

```python
# 거듭제곱
array_01 = np.array([1, 2, 3])
array_02 = np.array([1, 2, 3])

newArr = np.power(array_01, array_02)
print("거듭제곱: ", newArr)
```

# 소숫점 정렬

# numpy.select
- 조건식 list

```python
temp_arr = np.arange(10)
temp_arr
```

```python
temp_arr = np.arange(10)
cond_list = [temp_arr > 5, temp_arr < 2]

choice_list = [temp_arr * 2, temp_arr + 100]
np.select(cond_list,choice_list, default = temp_arr)
```

# NumPy와 Pandas와의 조합
- 어떤 과일 가게의 데이터셋을 임의로 생성했다.
- https://medium.com/analytics-vidhya/demystifying-np-where-and-np-select-76e3ca49e316 

```python
import pandas as pd 
import numpy as np

df = pd.DataFrame({
    '과일': ['사과', '자두', '사과', '자두'],
    '판매사': ['A', 'A', 'B', 'B'],
    '판매무게 (kg)': [10,20,30,40],
    '판매별점': [4, 3, 5, 3.5]
})
```

```python
A_price = pd.DataFrame({
    '과일': ['사과', '바나나', '자두', '파인애플'],
    '가격 (kg)': [7000, 5000, 5000, 6000]
})

B_price = pd.DataFrame({
    '과일': ['사과', '바나나', '자두', '파인애플'],
    '가격 (kg)': [6500, 5500, 6000, 5000]
})
```

- 기존 df에 각 품목에 맞는 것만 가져와서 kg당 가격을 추가하는 코드를 짜본다. 
- 먼저 `fruit` 함수를 인덱스로 추가한다. 

```python
df = df.set_index("과일")
A_price = A_price.set_index("과일")
B_price = B_price.set_index("과일")
```

- np.where 함수를 적용하여 각 판매사에 맞는 kg당 가격을 추가한다. 

```python
df['가격 (kg)'] = np.where(
    df['판매사'] == 'A', 
    A_price.loc[df.index]['가격 (kg)'], 
    B_price.loc[df.index]['가격 (kg)']
)

df
```

- 이번에는 판매사가 3개일 때 경우의 수를 생각해본다. 

```python
import pandas as pd 
import numpy as np

df = pd.DataFrame({
    '과일': ['사과', '자두', '사과', '자두', '사과', '자두'],
    '판매사': ['A', 'A', 'B', 'B', 'C', 'C'],
    '판매무게 (kg)': [10,20,30,40, 15, 20],
    '판매별점': [4, 3, 5, 3.5, 4, 5]
})

df_price = pd.DataFrame({
    '과일': ['사과', '바나나', '자두', '파인애플'],
    'A' : [7000, 5000, 5000, 6000], 
    'B' : [6500, 5500, 6000, 5000], 
    'C' : [2000, 3000, 3000, 4000]
})
```

```python
df = df.set_index("과일")
df
```

```python
df_price = df_price.set_index('과일')
df_price
```

```python
args = df_price.loc[df.index]
args
```

```python
conds = [
    df['판매사'] == 'A', 
    df['판매사'] == 'B', 
    df['판매사'] == 'C', 
]

choices = [
    args['A'], 
    args['B'], 
    args['C']
]
choices
```

```python
df['price (kg)'] = np.select(conds, choices)
df
```
