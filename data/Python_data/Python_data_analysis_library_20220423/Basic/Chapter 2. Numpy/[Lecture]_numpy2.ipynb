# 배열 생성의 다양한 방법들

- 모두 0으로 채운다. 

```python
# 0으로 채운다.
temp_arr = np.zeros((2, 3))
temp_arr
```

- 모두 1로 채운다. 

```python
# 
temp_arr = np.ones((3, 2, 2))
temp_arr
```

- 임의의 상수값으로 채운다. 

```python
# 
temp_arr = np.full((3, 2), fill_value=1.5)
temp_arr
```

- 이번에는 최소 및 최대 숫자의 범위를 정해두고 구간별로 값을 생성하는 코드를 작성한다. 

```python
#
temp_arr = np.linspace(5, 10, num = 30)
temp_arr
```

- 반복문 시, 자주 사용하는 배열을 작성한다. 

```python
#
temp_arr = np.arange(1, 11, 3)
temp_arr
```

- 이번에는 임의의 난수 생성을 확인한다. 

```python
#
import numpy as np
np.random.seed(42)
np.random.rand()
```

- 랜덤 배열을 생성한다. (Integer)




```python
# 
x = np.random.randint(100, size = (5))
print(x)
```

```python
#
x = np.random.randint(100, size = (2, 5))
print(x)
```

- 랜덤 배열을 생성한다. (Floats)

```python
# 
x = np.random.rand(2, 5)
print(x)
```

# 배열 연산
- 간단한 사칙 연산을 구현한다. 

```python
# 
array_01 = np.array([1,2,3])
array_02 = np.array([10,20,30])

array_01 + array_02
```

```python
#
newArr = np.add(array_01, array_02)
print("덧셈:", newArr)
```

```python
#
newArr = np.subtract(array_01, array_02)
print("뺄셈:", newArr)
```

```python
# 
newArr = np.multiply(array_01, array_02)
print("곱셈:", newArr)
```

```python
# 나눗셈
newArr = np.divide(array_01, array_02)
print("나눗셈:", newArr)
```

```python
# 거듭제곱
array_01 = np.array([1,2,3])
array_02 = np.array([1,2,3])

newArr = np.power(array_01, array_02)
print("거듭제곱:", newArr)
```

# 소숫점 정렬

- 소숫점을 제거하는 코드를 작성한다. 

```python
#
temp_arr = np.trunc([-1.23, 1.23])
print(temp_arr)
```

```python
#
temp_arr = np.fix([-1.23, 1.23])
print(temp_arr)
```

- 임의의 소숫점 자리에서 반올림

```python
# 
temp_arr = np.around([-1.2312314, 1.212343], 4)
print(temp_arr)
```

```python
#
temp_arr = np.around([-1.2312314, 1.212343], 4)
print(temp_arr)
```

- 소수점 모두 내린다. 

```python
#
temp_arr = np.floor([-1.2312314, 1.212343])
print(temp_arr)
```

- 소수점 모두 올린다. 

```python
#
temp_arr = np.ceil([-1.2312314, 1.212343])
print(temp_arr)
```

# numpy.where
- 조건식

```python
# 
temp_arr = np.arange(10)
temp_arr
# 5미만 0, 1, 2, 3, 4
# 5이상 * 10
np.where(temp_arr < 5, temp_arr, temp_arr * 10)
```

# numpy.select
- 조건식 list

```python
temp_arr = np.arange(10)
temp_arr

conditional_list = [temp_arr > 5, temp_arr < 3]
result_list      = [temp_arr,     temp_arr + 100]

np.select(conditional_list, result_list, default = 99999)
```
