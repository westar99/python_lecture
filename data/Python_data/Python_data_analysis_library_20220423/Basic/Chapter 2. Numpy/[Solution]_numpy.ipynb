## 사칙 연산
- 이번에는 임의의 수학 점수와 영어 점수를 합산하는 코드를 작성한다. 
- 그런데, List를 연산하면 원하는 결괏값이 나타나지 않는다. 
- 따라서, 리스트를 배열로 변환해야 한다. 

```python
math_scores    = [90, 80, 88, 75, 85]
english_scores = [90, 100, 88, 80, 85]

math_scores + english_scores
```

```python
math_scores = [90, 80, 88, 75, 85]
english_scores = [90, 100, 88, 80, 85]

math_array = np.array(math_scores)
english_array = np.array(english_scores)

total_scores = math_array + english_array
```

- 최소 및 최대 값을 구한다. 

```python
total_scores
```

```python
np.min(total_scores)
```

```python
np.max(total_scores)
```

## 배열의 생성
- 배열은 크게 0차원부터 3차원까지로 구성한다. 

```python
temp_arr = np.array(30)
print(type(temp_arr))
print(temp_arr.shape)
```

```python
# 1차원
temp_arr = np.array([1, 2, 3])
print(type(temp_arr))
print(temp_arr.shape)
print(temp_arr.ndim)
```

```python
# 2차원
temp_arr = np.array([[1, 2, 3], [4, 5, 6]])
print(type(temp_arr))
print(temp_arr.shape)
print(temp_arr.ndim)
```

```python
# 3차원 --> 이미지 
temp_arr = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])
print(type(temp_arr))
print(temp_arr.shape)
print(temp_arr.ndim)
```

```python
import numpy as np

arr = np.array([1, 2, 3, 4], ndmin=2)

print(arr)
print('shape of array :', arr.shape)
```
